```python
import geopandas as gpd
import matplotlib.pyplot as plt


data = gpd.read_file("PE_Municipios_2021.shp")
display(data)
```

```python
for c in range(0,len(data.index)):
    data.iat[c,1]=data.iat[c,1].lower()
display(data)
```

```python
data.plot(figsize=(16,14),facecolor='white',edgecolor= 'black')
plt.show()
```

```python
dado_recife = data[data['NM_MUN']=='Recife']

dado_recife.plot(facecolor='white',edgecolor= 'black')
plt.show()
```
